```python
class svd_pca():
    '''calculating eigen vector using power method and sgd. User provides which method to use as input
    
    Parameters
    ----------------
    n_component: int
        number of principle components to calculate. They are in the directions of maximum variance of data.
    
    Attributes
    ----------------
    eigvec_: array, shape (n_components, xtrain.shape[1])
        principle components explaining maximum variance. 
        
    eigval_: array, shape (n_components,)
        variance in direction of each principle component.
    '''
    
    def __init__(self, n_component):
        self.n_component = n_component
        self.eigvec_ = 0
        self.eigval_ = 0 
    
    def svd_fit(self, xtrain, method='gd'):
        '''This function finds eigenvector and calculates eigenvalues, and is called
        number of times n_components is set to.
        
        Parameters
        ----------------
        xtrain: pd.dataframe
            
        method: str (default 'gd') {'gd', 'power'}
            specify which algorithm to use for calculating the eigenvector. 
        '''
        
        x = xtrain.copy()
        svd_list=[]
        if type(self.n_component) is float:
            raise TypeError('n_components should be integer greater than 0')
        for i in range(self.n_component):
            if (method is 'power'): 
                eig_vec = self.power_eigen_vec(x) #run this for PCA power method
            elif (method is 'gd'):
                eig_vec = self.gd_eigen_vec(x) #run this for PCA gd
            else:
                print('This method is not implemented')
            #calculating eigenvalue using cov matrix of input data
            eig_val = eig_vec.dot(x.cov()).dot(eig_vec)
            svd_list.append((eig_val, eig_vec))
            #subtracting Rank1 matrix from original matrix to get the next orthogonal vector 
            x -= (x.dot(eig_vec)[:,None])*eig_vec
        self.eigval_, self.eigvec_ = np.array(list(zip(*svd_list))[0]), np.array(list(zip(*svd_list))[1])
    
    def power_eigen_vec(self, xtrain, threshold = 1e-05):
        ''' This function calculates eigenvectors by iterative power method, Lanczos algorithm) 
        '''
        x_square = np.dot(xtrain.T, xtrain)
        vec = np.random.rand(x_square.shape[1])
        vec_norm = vec/np.linalg.norm(vec)
        for i in range(1000):
            u_vec = np.dot(x_square, vec_norm)
            u_vec_norm = u_vec/np.linalg.norm(u_vec)
            error = np.dot(u_vec_norm, vec_norm)
            vec_norm = u_vec_norm

            if (abs(error) > 1 - threshold):
                print(i) #printing i just to observe no of iterations 
                return vec_norm 
        return vec_norm
    
    def gd_eigen_vec(self, X, l_rate = 0.01):
        ''' This function calculates eignevectors by finding direction that maximize variance of the data. Gradient 
        descent is used to converge faster to eigenvector.
        '''
        X_demean = X.values -np.mean(X).values
        ran_vec = np.random.rand(X.shape[1])
        vec_norm = ran_vec/np.linalg.norm(ran_vec)
        variance = (X_demean.dot(vec_norm)).T.dot(X_demean.dot(vec_norm))/X_demean.shape[0]
        update_epoch = [1]
        i = 0
        for epoch in range(50):
            #x = random.sample(X_demean, 10)
            i +=1
            w = vec_norm
            old_variance = variance
            variance_epoch = (X_demean.dot(w)).T.dot(X_demean.dot(w))/X_demean.shape[0] 
            variance += variance_epoch
            #w = w +l_rate*(row.dot(w))*row
            update = (l_rate*(X_demean.T.dot(X_demean.dot(w))*2 -2*w))
            update_epoch.append(np.linalg.norm(update))
            vec_norm = w + update
            vec_norm = vec_norm/np.linalg.norm(vec_norm)

            if (abs((update_epoch[-1]) - update_epoch[-2])) < 1e-5:
                print(i) #printing i just to observe no of iterations 
                return vec_norm
        return (vec_norm)


    def transform_eigen_subspace(self, X):
        '''This function create linear transformation of the original dataset matrix into eigenvector subspace. The 
        dataset has to be centered first, before trasformation. V.dot(V.T) is identity matrix as V is a set of orthonormal
        basis vector formed by eigenvectors. Also according to linear transformation, A.dot(orthonormal x-y basis vector)= 
        (transformed_A).dot(orthonormal eigenvector). hence transformed_A = A.dot(orthonormal x-y basis vector).(orthonormal
        eigenvector) inverse. inverse of V = V.T.
        '''
        X_demean = X.values -np.mean(X).values
        #I am using only fisrt 4 vectors to transform the data. 
        eigen_vec_matrix = np.array(self.eigvec_[:4]).T
        X_transform = X_demean.dot(eigen_vec_matrix)
        return X_transform

    def inverse_transform(self, X_transformed):
        eigen_vec_matrix = np.array(self.eigvec_[:4])
        #print(eigen_vec_matrix)
        inverse_transform = (X_transformed.values.dot(eigen_vec_matrix))
        return inverse_transform
        
```
