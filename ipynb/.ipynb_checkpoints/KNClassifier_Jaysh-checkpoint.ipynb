```python
titanic_df = pd.read_csv('../Data/titanic.csv')
```

```python
titanic_df.set_index('PassengerId', inplace=True)
titanic_df.head()
```

### Setting the target and predictor variables

```python
target = titanic_df['Survived']
```

```python
predictor = titanic_df[['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Embarked']]
```

```python
predictor.isnull().sum()
```

```python
predictor['Age'].fillna(np.mean(predictor['Age']), inplace= True)
```

```python
predictor.Embarked.unique()
```

```python
predictor.Embarked.fillna(mode(predictor.Embarked), inplace=True)
```

```python
predictor = pd.get_dummies(predictor)
```

```python
x_train, x_test, y_train, y_test = train_test_split(predictor, target, test_size = 0.3)
```

```python
x_train.shape, x_test.shape
```

```python
y_train.value_counts()
```

```python
from scipy.spatial.distance import cdist

class Knn_Classifier():
    '''Class implements KNeighbors classifier by measuring pairwise euclidean distance between each point in dataset. 
    For every k neighbors with minimun distance found, the class label is decided based on majority class.
    
    Parameters
    ----------------
    neighbors: int
        number of nearest neighbors to be considered for decided class label. 
        
    '''
    
    def __init__(self, neighbors = 3):
        self.neighbors = neighbors

    def fit(self, x,y):
        pass
    
    def predict(self, x, y):
        
        sorted_distance = []
        for i in x.values:
            i = i.reshape(-1, i.shape[0])
            euclidean_dist = cdist(x.values, i)
            #euclidean_dist= euclidean_dist.reshape(euclidean_dist.shape[1], -1)
            euclidean_index = list(zip(euclidean_dist, x.index)) 
            sorted_distance.append(sorted(euclidean_index))
                #distance.append([euclidean_dist, index])
            #sorted_distance.append(sorted(distance))
       # print(sorted_distance)

        prediction =[]

        for i in sorted_distance:
            target = []
            for j in range(self.neighbors):
                index = i[j][1]
                target.append(y[index])
            #print(target)

            a = (Counter(target).most_common(1))[0][0] #replace with np.mean for knregressor
   
    #         print(a)
            prediction.append(a)
        return prediction
    
    def acc_score(self, x, y):
        predicted = []
        predicted = self.predict(x, y)
        #print(len(y_test), len(predicted))
        TP = np.sum(np.logical_and(np.array(y) == 1, np.array(predicted) == 1))
        TN = np.sum(np.logical_and(np.array(y) == 0, np.array(predicted) == 0))
        FP = np.sum(np.logical_and(np.array(y) == 0, np.array(predicted) == 1))
        FN = np.sum(np.logical_and(np.array(y) == 1, np.array(predicted) == 0))
        return float(TP + TN)/float(TP + TN + FP + FN)
```

```python
knn = Knn_Classifier(neighbors=13)
knn.fit(x_train, y_train)
```

```python
knn.predict(x_train, y_train)
```

```python
knn.acc_score(x_train, y_train)
```

I can improve the prediction and score by adding weights to the neighbors. In that algorithms, the nearest neighbor will get more weight in prediction compared to far away neighbors.

```python
from scipy.spatial import distance
```
